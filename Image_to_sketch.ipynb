IMAGE To Pencil sketch app


```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import Image as display
```

```python
loc='/content/BIRD.jpeg'
image = Image.open(loc)
print(image.size)
image1 = cv2.imread(loc)
cv2.imwrite(loc, image1)
display(filename=loc)
```

```python
img = cv2.imread(loc)
grayed = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(grayed)
cv2.imwrite(loc, grayed)
display(filename=loc)
```

```python
inverted = cv2.bitwise_not(grayed)
cv2.imwrite("inverted.png", inverted)
display(filename="inverted.png")
```

```python
blurred = cv2.GaussianBlur(inverted, (15, 15), sigmaX=0, sigmaY=0)
cv2.imwrite("blurred.png", blurred)
display(filename="blurred.png")
```

```python
#Pencil sketch image
```

```python
def blend(x, y):
    return cv2.divide(x, 255 - y, scale=257)
final_result = blend(grayed, blurred)
cv2.imwrite("final.png", final_result)
display(filename="final.png")
```
